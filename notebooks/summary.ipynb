```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage.transform import resize 

stone_paths = [
    "/home/stevexu/VSprojects/ELITE/assets/images/stone.png",
    "/home/stevexu/VSprojects/ELITE/outputs/local_mapping_with_ipadapter/stone_scl0p8_l0p8/00003_00042.jpg",
    "/home/stevexu/VSprojects/ELITE/outputs/local_mapping_with_ipadapter/stone_scl0p8_l0p8/00010_00042.jpg",
]  

statue_paths = [
    "/home/stevexu/VSprojects/ELITE/assets/images/statue.png",
    "/home/stevexu/VSprojects/ELITE/outputs/local_mapping_with_ipadapter/statue_scl0p8_l0p8/00004_00042.jpg",
    "/home/stevexu/VSprojects/ELITE/outputs/local_mapping_with_ipadapter/statue_scl0p8_l0p8/00005_00042.jpg",
]

vermeer_paths = [
    "/home/stevexu/VSprojects/ELITE/control_modules/control_imgs/images/vermeer.jpg",
    "/home/stevexu/VSprojects/ELITE/outputs/local_mapping_with_ipadapter/vermeer_scl0p8_l0p8/00003_00042.jpg",
    "/home/stevexu/VSprojects/ELITE/outputs/local_mapping_with_ipadapter/vermeer_scl0p8_l0p8/00001_00042.jpg",
]

img_paths = stone_paths + statue_paths + vermeer_paths
images = [] 
for path in img_paths:
    img = mpimg.imread(path)
    if img.shape[1] == 2*img.shape[0]:
        img1 = img[:, :img.shape[0], :]
        img2 = img[:, img.shape[0]:, :]
        images.append(img2)
        images.append(img1)
        continue
    images.append(img)

fig, axs = plt.subplots(3, 5, figsize=(10, 6))

titles = ['Input of Ip-Adapter', 'the object', 'output', 'the object', 'output']  # Replace with your actual titles

for i, ax in enumerate(axs.flatten()):
    img = images[i]  # Read the image file into a numpy array
    # print(img.shape)
    ax.imshow(img)
    ax.axis('off')  # to hide the axis
    if i < len(titles):  # Add title to the first row
        ax.set_title(titles[i])

plt.show()
```
